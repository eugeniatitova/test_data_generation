```python
import pandas as pd
import numpy as np
import math

from scipy.stats import entropy

import warnings
warnings.filterwarnings("ignore")
```

```python
original_df = pd.read_csv("heart.csv")
```

```python
generated_df = pd.read_csv("generated.csv", header=None)
generated_df.columns = original_df.columns
```

```python
from sklearn.metrics import mutual_info_score 

def I(x, y): 
    mi = mutual_info_score(x, y) 
    return mi 
```

```python
x = original_df.loc[:100, :].values.flatten()
```

```python
entropy(pd.Series(x).value_counts())
```

```python
y = generated_df.loc[:100, :].values.flatten()
```

```python
entropy(pd.Series(y).value_counts())
```

```python
I(x, y)
```
